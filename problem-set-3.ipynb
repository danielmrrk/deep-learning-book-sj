##### Problem 3.5

Prove that the following property holds for α ∈ R+:  
ReLU[α · z] = α · ReLU[z].  
This is known as the non-negative homogeneity property of the ReLU function.  



**Given:**
$
\alpha \in \mathbb{R}^+
$

**Statement:**
$
\text{ReLU}[\alpha \cdot z] = \alpha \cdot \text{ReLU}[z]
$

**Proof:**

$
\text{ReLU}[\alpha \cdot z] = 
\begin{cases} 
\alpha \cdot z, & \text{if } \alpha \cdot z \geq 0 \\
0, & \text{if } \alpha \cdot z < 0 
\end{cases}
$

$\text{Since } \alpha \in \mathbb{R}^+ \text{ and  z in } \mathbb{R}\,$

$
\text{ReLU}[z] = 
\begin{cases} 
z, & \text{if } z \geq 0 \\
0, & \text{if } z < 0 
\end{cases}
$

$\text{Thus, for z }\geq \text{0, we have:}$

$
\begin{aligned}
&z >= 0 \Leftrightarrow \alpha \cdot z \geq 0 \\
&\Rightarrow \text{ReLU}[\alpha \cdot z] \\
&= \alpha \cdot z \\
&= \alpha \cdot \text{ReLU}[z]
\end{aligned}
$

$\text{For z < 0, we have:}$

$
\begin{aligned}
&\alpha \cdot z < 0 \\
&\Rightarrow \text{ReLU}[\alpha \cdot z] = 0 \\
&= \alpha \cdot \text{ReLU}[z]
\end{aligned}
$

$\text{Therefore, in both cases, }\text{ReLU}[\alpha \cdot z] = \alpha \cdot \text{ReLU}[z].$

$\text{It's very important to understand that this only holds since }\alpha \in \mathbb{R}^+.$

##### Problem 3.6

Following on from problem 3.5, what happens to the shallow network defined in  
equations 3.3 and 3.4 when we multiply the parameters θ10 and θ11 by a positive constant α  
and divide the slope ϕ1 by the same parameter α? What happens if α is negative?

**Given:**

$
h_1 = a[\alpha \cdot (\theta_0 + \theta_1 x)] = \alpha \cdot \sigma[\theta_0 + \theta_1 x]
$

**Let:**

$
\begin{aligned}
&\text{If } \alpha \in \mathbb{R}^+ \text{ then we can use the non-negative homogeneity rule}\\
&y = \phi_0 + \frac{\phi_1}{\alpha} \alpha \cdot a[\theta_0 + \theta_1 x] \\
&\text{Which then leads to the fact that the alphas cancel out.} \\ \\
\end{aligned}
$


$
\text{Now we consider the cases where } \alpha \in \mathbb{R}^-.
$

**Case 1:** 
$\text{If } z = \theta_0 + \theta_1 x < 0 \text{ which then will lead to } \alpha \cdot z > 0
$

$
\begin{aligned}
&\text{Therefore } a[\alpha \cdot z] = \alpha \cdot z. \\
&\text{Later the alphas will be canceled out.} \\ \\
\end{aligned}
$

**Case 2:**
$\text{If } z = \theta_0 + \theta_1 x \geq 0 \text{ which then will lead to } \alpha \cdot z \leq 0
$

$
\begin{aligned}
&\text{Therefore } a[\alpha \cdot z] = 0 \\ \\
\end{aligned}
$

**Results:**

$
\text{When } \alpha \in \mathbb{R}^+ \text{ this won't change anything since we have this non-negative homogeneity rule } \\
\text{and therefore the alpha will simply cancel out. On the other side when we have } \alpha \in \mathbb{R}^- \text{ then}\\
\text{this will lead to an inversion of clipping, which means that now the positive side of the linear function} \\
\text{will be clipped and the negative won't. }
\text{But in the end this will only do this inversion and not} \\
\text{more since the alphas cancel here out as well.}
$





```python
from matplotlib import pyplot as plt
import numpy as np


def relu(x):
    return np.maximum(0, x)

alpha = 2
phi = 2 / alpha
x = np.linspace(-100, 100, 100)
y = phi * x * alpha
fig, ax = plt.subplots(nrows = 1, ncols= 3, figsize=(10, 4))

ax[0].plot(x, y)
ax[0].grid(True)
ax[0].set_title("No Activation Function")
ax[0].set_xlim(-100, 100)
ax[0].set_ylim(-100, 100)

# Plot ReLU function
y_pos_alpha = phi * relu(alpha * x)
ax[1].plot(x, y_pos_alpha, color='orange')
ax[1].grid(True)
ax[1].set_xlim(-100, 100) 
ax[1].set_ylim(-100, 100)

y_pos_alpha = phi * relu(alpha * x)
ax[1].plot(x, y_pos_alpha, color='red')
ax[1].grid(True)
ax[1].set_xlim(-100, 100) 
ax[1].set_ylim(-100, 100)


plt.show()
```
