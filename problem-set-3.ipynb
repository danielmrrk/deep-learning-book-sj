##### Problem 3.5

Prove that the following property holds for α ∈ R+:  
ReLU[α · z] = α · ReLU[z].  
This is known as the non-negative homogeneity property of the ReLU function.  



**Given:**
$
\alpha \in \mathbb{R}^+
$

**Statement:**
$
\text{ReLU}[\alpha \cdot z] = \alpha \cdot \text{ReLU}[z]
$

**Proof:**

$
\text{ReLU}[\alpha \cdot z] = 
\begin{cases} 
\alpha \cdot z, & \text{if } \alpha \cdot z \geq 0 \\
0, & \text{if } \alpha \cdot z < 0 
\end{cases}
$

$\text{Since } \alpha \in \mathbb{R}^+ \text{ and z} \in \mathbb{R}\,$

$
\text{ReLU}[z] = 
\begin{cases} 
z, & \text{if } z \geq 0 \\
0, & \text{if } z < 0 
\end{cases}
$

$\text{Thus, for z }\geq \text{0, we have:}$

$
\begin{aligned}
&z >= 0 \Leftrightarrow \alpha \cdot z \geq 0 \\
&\Rightarrow \text{ReLU}[\alpha \cdot z] \\
&= \alpha \cdot z \\
&= \alpha \cdot \text{ReLU}[z]
\end{aligned}
$

$\text{For z < 0, we have:}$

$
\begin{aligned}
&\alpha \cdot z < 0 \\
&\Rightarrow \text{ReLU}[\alpha \cdot z] = 0 \\
&= \alpha \cdot \text{ReLU}[z]
\end{aligned}
$

Therefore, in both cases, $\text{ReLU}[\alpha \cdot z] = \alpha \cdot \text{ReLU}[z].$

It's very important to understand that this only holds since $\alpha \in \mathbb{R}^+.$

##### Problem 3.6

Following on from problem 3.5, what happens to the shallow network defined in  
equations 3.3 and 3.4 when we multiply the parameters θ10 and θ11 by a positive constant α  
and divide the slope ϕ1 by the same parameter α? What happens if α is negative?

**Given:**

$
h_1 = a[\alpha \cdot (\theta_0 + \theta_1 x)] = \alpha \cdot a[\theta_0 + \theta_1 x]
$

**Let:**

$
\begin{aligned}
&\text{If } \alpha \in \mathbb{R}^+ \text{ then we can use the non-negative homogeneity rule}\\
&y = \phi_0 + \frac{\phi_1}{\alpha} \alpha \cdot a[\theta_0 + \theta_1 x] \\
&\text{Which then leads to the fact that the alphas cancel out.} \\ \\
\end{aligned}
$


$
\text{Now we consider the cases where } \alpha \in \mathbb{R}^-.
$

**Case 1:** 
$\text{If } z = \theta_0 + \theta_1 x < 0 \text{ which then will lead to } \alpha \cdot z > 0
$

$
\begin{aligned}
&\text{Therefore } a[\alpha \cdot z] = \alpha \cdot z. \\
&\text{Later the alphas will be canceled out.} \\ \\
\end{aligned}
$

**Case 2:**
$\text{If } z = \theta_0 + \theta_1 x \geq 0 \text{ which then will lead to } \alpha \cdot z \leq 0
$

$
\begin{aligned}
&\text{Therefore } a[\alpha \cdot z] = 0 \\ \\
\end{aligned}
$

**Results:**


When $\alpha \in \mathbb{R}^+$  this won't change anything since we have this non-negative homogeneity rule  
and therefore the alpha will simply cancel out. On the other side when we have $\alpha \in \mathbb{R}^-$ then  
this will lead to an inversion of clipping, which means that now the positive side of the linear function  
will be clipped and the negative won't. But in the end this will only do this inversion and not  
more since the alphas cancel here out as well.






```python
from matplotlib import pyplot as plt
import numpy as np


def relu(x):
    return np.maximum(0, x)

alpha = 2
phi = 2 / alpha
x = np.linspace(-100, 100, 100)
y = phi * x * alpha
fig, ax = plt.subplots(nrows = 1, ncols= 3, figsize=(10, 4))

ax[0].plot(x, y)
ax[0].grid(True)
ax[0].set_title("No Activation Function")
ax[0].set_xlim(-100, 100)
ax[0].set_ylim(-100, 100)

# Plot ReLU function
y_pos_alpha = phi * relu(alpha * x)
ax[1].plot(x, y_pos_alpha, color='orange')
ax[1].grid(True)
ax[1].set_title("ReLU with pos. alpha")
ax[1].set_xlim(-100, 100) 
ax[1].set_ylim(-100, 100)

y_pos_alpha = -phi * relu(-alpha * x)
ax[2].plot(x, y_pos_alpha, color='red')
ax[2].set_title("ReLU with neg. alpha")
ax[2].grid(True)
ax[2].set_xlim(-100, 100) 
ax[2].set_ylim(-100, 100)


plt.show()
```

Consider replacing the ReLU activation function with (i) the Heaviside step function
heaviside[z], (ii) the hyperbolic tangent function tanh[z], and (iii) the rectangular function
rect[z], where:
heaviside[z] =


$
\text{heaviside}[z] = 
\begin{cases} 
0 & z < 0 \\
1 & z \geq 0 
\end{cases}
$

$
\text{rect}[z] = 
\begin{cases} 
0 & z < 0 \\
1 & 0 \leq z \leq 1 \\
0 & z > 1 
\end{cases}
\quad (3.15)
$

Redraw a version of figure 3.3 for each of these functions. The original parameters were: ϕ =  
{ϕ0, ϕ1, ϕ2, ϕ3, θ10, θ11, θ20, θ21, θ30, θ31} = {−0.23,−1.3, 1.3, 0.66,−0.2, 0.4,−0.9, 0.9, 1.1,−0.7}.  
Provide an informal description of the family of functions that can be created by neural networks  
with one input, three hidden units, and one output for each activation function.  

```python
from functools import partial

def heaviside(z):
    if z < 0:
        return 0
    return 1

def rect(z):
    if z < 0 or z > 1:
        return 0
    return 1

def hidden_cell(x, theta_0, theta_1, activation):
    z = theta_0 + theta_1 * x
    output = np.array([activation(z_value) for z_value in z])
    return z, output

x = np.linspace(0, 2, 100)

h1 = partial(hidden_cell, x=x, theta_0=-0.2, theta_1=0.4)
h2 = partial(hidden_cell, x=x, theta_0=-0.9, theta_1=0.9)
h3 = partial(hidden_cell, x=x, theta_0=1.1, theta_1=-0.7)

activation_functions = {
    heaviside: "Heaviside",
    rect: "Rect",
    np.tanh: "Tanh"
}
```

```python

phi_0 = -0.23
phi_1 = -1.3
phi_2 = 1.3
phi_3 = 0.66


def plot_shallow_neuronal_network_progress(activation):
    fig, ax = plt.subplots(nrows=3, ncols= 3, figsize=(10,10))

    fig.suptitle(f'Activation Function: {activation_functions[activation]}', fontsize=16)

    z, output = h1(activation=activation)
    ax[0][0].plot(x, z, label=r"$\theta_{10} + \theta_{11}x$", color="red")
    ax[0][0].set_xlim(0, 2)
    ax[0][0].set_ylim(-1, 1)
    ax[0][0].legend()

    ax[1][0].plot(x, output, label=r"$h1 = a[\theta_{10} + \theta_{11}x]$", color="red")
    ax[1][0].set_xlim(0, 2)
    ax[1][0].set_ylim(-1, 1)
    ax[1][0].legend()

    weighted_h1 = phi_1 * output
    ax[2][0].plot(x, weighted_h1, label=r"$\phi_1h1$", color="red")
    ax[2][0].set_xlim(0, 2)
    ax[2][0].set_ylim(-1, 1)
    ax[2][0].legend()

    z, output = h2(activation=activation)
    ax[0][1].plot(x, z, label=r"$\theta_{20} + \theta_{21}x$", color="blue")
    ax[0][1].set_xlim(0, 2)
    ax[0][1].set_ylim(-1, 1)
    ax[0][1].legend()

    ax[1][1].plot(x, output, label=r"$h2 = a[\theta_{20} + \theta_{21}x]$", color="blue")
    ax[1][1].set_xlim(0, 2)
    ax[1][1].set_ylim(-1, 1)
    ax[1][1].legend()

    weighted_h2 = phi_2 * output
    ax[2][1].plot(x, weighted_h2, label=r"$\phi_2h2$", color="blue")
    ax[2][1].set_xlim(0, 2)
    ax[2][1].set_ylim(-1, 1)
    ax[2][1].legend()

    z, output = h3(activation=activation)
    ax[0][2].plot(x, z, label=r"$\theta_{30} + \theta_{31}x$", color="black")
    ax[0][2].set_xlim(0, 2)
    ax[0][2].set_ylim(-1, 1)
    ax[0][2].legend()

    ax[1][2].plot(x, output, label=r"$h2 = a[\theta_{30} + \theta_{31}x]$", color="black")
    ax[1][2].set_xlim(0, 2)
    ax[1][2].set_ylim(-1, 1)
    ax[1][2].legend()

    weighted_h3 = phi_3 * output
    ax[2][2].plot(x, weighted_h3, label=r"$\phi_3h3$", color="black")
    ax[2][2].set_xlim(0, 2)
    ax[2][2].set_ylim(-1, 1)
    ax[2][2].legend()

    for i in range(3):
        ax[i][1].yaxis.set_visible(False)
        ax[i][2].yaxis.set_visible(False)

    plt.show()

    return weighted_h1, weighted_h2, weighted_h3

```

```python
def plot_activation_output(weighted_h1, weighted_h2, weighted_h3):
    y_hat = phi_0 + weighted_h1 + weighted_h2 + weighted_h3
    plt.plot(x, y_hat, label=r"$\phi_0 + \phi_1h1 + \phi_2h2 + \phi_3h3$")
    plt.legend()
    plt.show()
```

```python
weighted_h1, weighted_h2, weighted_h3 = plot_shallow_neuronal_network_progress(heaviside)
plot_activation_output(weighted_h1, weighted_h2, weighted_h3)
```

```python
weighted_h1, weighted_h2, weighted_h3 = plot_shallow_neuronal_network_progress(rect)
plot_activation_output(weighted_h1, weighted_h2, weighted_h3)
```

```python
weighted_h1, weighted_h2, weighted_h3 = plot_shallow_neuronal_network_progress(np.tanh)
plot_activation_output(weighted_h1, weighted_h2, weighted_h3)
```

$x_1=-\frac{\theta_{10}}{\theta_{11}}, x_2=-\frac{\theta_{20}}{\theta_{21}}, x_3=-\frac{\theta_{30}}{\theta_{31}}$

$J := \{x_1, x_2, x_3\}$

Or simply said when the clipping or "joint" x-positions are the same, then there will be less linear regions
